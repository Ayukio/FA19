```python
"""реализовать решение квадратного уравнения через дискриминант"""
import math
def is_digital(input_str):
        try:
            float(input_str)
            return True
        except:
            return False

def parsing(input_str):
    
        global coeff_a, coeff_b, coeff_c
        symbols = 1
        find = False
        while find == False:
            
            coeff_a = input_str[:symbols]
            if is_digital(coeff_a) == False and symbols != 1:
                find = True

            symbols += 1

        coeff_a = input_str[:symbols-2]

        
        symbols = input_str.rindex("x")
        before_symbol_x = input_str[:symbols]
        symbols = 1
        find = False
        while find == False:

            coeff_b = before_symbol_x[symbols:]            

            if is_digital(coeff_b) == True:
                find = True
            
            symbols += 1


        symbols = input_str.rindex("=")
        before_symbol_equality = input_str[:symbols]
        symbols = 1
        find = False
        while find == False:

            coeff_c = before_symbol_equality[symbols:]            

            if is_digital(coeff_c) == True:
                find = True
            
            symbols += 1

        print("a =",coeff_a)
        print("b =",coeff_b)
        print("c =",coeff_c)

        coeff_a = float(coeff_a)
        coeff_b = float(coeff_b)
        coeff_c = float(coeff_c)
def equation_solver(input_str):      
        a = coeff_a
        b = coeff_b
        c = coeff_c

        D = b ** 2 - 4 * a * c
        print("\nD =",D)

        if D == 0:
            x1 = -b/(2*a)
            print("Ответ:")
            print("x1 =",x1)

        elif D > 0:
            x1 = (-b + math.sqrt(D)) / (2 * a)
            x2 = (-b - math.sqrt(D)) / (2 * a)
            print("Ответ:")
            print("Корень x1 =",x1)
            print("Корень x2 =",x2)
        
        else:
            print("Нет решения, D < 0")

start = 1
while start == 1:
       
    input_str = input("Введите квадратное уравнение вида ax^2+bx+c=0: ")
    if input_str[0] =="-" or input_str[0] =="+"  and input_str[1] =="0":
        print( "Коэф 'a' равен нулю")
        continue
    elif input_str[0] == "0":
        print( "Коэф 'a' равен нулю!")
        continue
    else:
        start = 0
parsing(input_str)
equation_solver(input_str)
```

```python
"""реализовать создание, запись, чтение и удаление файла с данными о пользователе."""
import os

print("1 - create a file")
print("2 - add a text in file")
print("3 - read a file")
print("4 - remove a file")
print("0 - exit\n")
oper = ""
filename = input("Enter name of your file: ")
while oper != "0":
    oper = input("Make a choice between 0-4 types: ")
    
    if oper == "1":
        #filename = input("Enter name of your file: ")
        my_file = open(filename, mode = "w")
        print("File with name " + filename + " has been created")
        my_file.close()
    elif oper == "2":
        #filename = input("Filename: ")
        text = input("Text: ")
        my_file = open(filename, "a")
        my_file.write(text)
        my_file.close()
    elif oper == "3":
        #filename = input("Filename: ")
        try:
            my_file = open(filename, "r")
            print("Text from file:\n")
            print(my_file.read())
            my_file.close()
        except FileNotFoundError:
            print("No such file or directory")

    elif oper == "4":
        #filename = input("Filename: ")
        try:
            os.remove(filename)
            print("The file is removed")
        except (OSError, NameError):
            print("File doesn't exist")
    elif oper == "0":
        break
    else:
        print("Enter 0-4 types")
```
