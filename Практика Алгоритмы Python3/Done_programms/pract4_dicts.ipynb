### Задача5
Создать словарь авиарейсов, с возможностью поиска маршрута из точки А в точку В с учетом одной пересадки.

```python
d = {"Москва": ("Лондон", "Белгород", "Каир"),
     "Каир" : ("Калининград", "Москва"),
     "Белгород": ("Москва", "Калининград"),
     "Калининград": ("Каир","Белгород"),
     "Лондон": ("Москва")
    }
city1 = input("Первый город: ").title()
city2 = input("Второй город: ").title()
find = False

if city1 in d:
    for el_1 in d[city1]:
        if el_1 == city2:
            find = True
            print("Прямой рейс:", city1, "—>", el_1)
            break

        for el_2 in d[el_1]: 
                if el_2 == city2:
                    find = True
                    print("Рейс:\n", city1, "—>", el_1, "\n", el_1, "—>", el_2)

    if find == False:
        print("По вашему запросу ничего не найдено")
else:
    print("Город не найден!")
```
